```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")
model = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC")
```

```python
sentence = "유한양행-더욱 탄탄해지는 R&D, 2022년 연결 매출액 1조 8,514억원(+9.7% yoy), 영업이익 651억원(+33.9% yoy, OPM 3.5%) 전망, 레이저티닙, 6월 ASCO학회 아미반타맙 병용 투여 임상 데이터 업데이트, 6월 AOS학회 단독 2차 치료제 및 3분기 단독 1차 치료제 임상 결과 발표 예정"
```

```python
inputs = tokenizer(
                    sentence,
                    return_tensors='pt',
                    truncation=True,
                    max_length=128,
                    padding='max_length',
                    add_special_tokens=True
)

output = model(input_ids=inputs['input_ids'],
                attention_mask=inputs['attention_mask'],
)
```

```python
output[0][0]
```

```python
import torch.nn as nn

m = nn.Softmax(dim=1)
result = m(output[0])
```

```python
negative_val = float(result[0][0])
neutral_val = float(result[0][1])
positive_val = float(result[0][2])
print("Negative :", negative_val, "\nNeutral  :", neutral_val, "\nPositive :", positive_val)
```
